```python
df = pd.read_csv("/kaggle/input/social-network-ads/Social_Network_Ads.csv")
```

```python
df.drop(columns=["User ID","Gender"],inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df.head()
```

```python
X= df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
scaler = StandardScaler()
```

```python
X_scale = scaler.fit_transform(X)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_scale,y,test_size=0.2,random_state=2)
```

```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(10,activation="relu",input_dim=2))
model.add(Dense(10,activation="relu"))
model.add(Dense(1,activation="sigmoid"))

```

```python
model.summary()
```

```python
#Stochastic Gradient Descent -->Batch_size ==1

model.compile(loss="binary_crossentropy",metrics=["accuracy"])
records = model.fit(X_train,y_train,epochs=50,batch_size=1)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred = np.where(y_pred>0.5,1,0)
```

```python
accuracy_score(y_test,y_pred)
```

```python
#Batch Gradient Descent -->Batch_size == no of rows in the dataset

model2 = Sequential()
model2.add(Dense(10,activation="relu",input_dim=2))
model2.add(Dense(10,activation="relu"))
model2.add(Dense(1,activation="sigmoid"))

model2.compile(loss="binary_crossentropy",metrics=["accuracy"])
records = model2.fit(X_train,y_train,epochs=50,batch_size=367)
```

```python
y_pred = model2.predict(X_test)
y_pred = np.where(y_pred>0.5,1,0)
accuracy_score(y_test,y_pred)
```
